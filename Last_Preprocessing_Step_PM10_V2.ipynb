```python
new_demo = demographic_data.iloc[:, :5]
data_with_demo = pd.merge(data, new_demo, on="LMB0A_IDcu", how='left')
```

```python
pd.set_option('display.max_rows', 20)
data_with_demo
```

```python
data_with_demo.replace(np.nan, 0, inplace=True)
data_with_demo
```

```python
for i in range(len(data_with_demo)):
    for j in range(2, 55):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] * data_with_demo['POP_2013'][i]
    for j in range(55, 108):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] * data_with_demo['POP_2014'][i]
    for j in range(108, 161):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] * data_with_demo['POP_2015'][i]
    for j in range(161, 214):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] * data_with_demo['POP_2016'][i]
        
data_with_demo
```

```python
data_with_demo.drop(['LMB0A_IDcu'], axis=1, inplace=True)
data_with_demo = data_with_demo.groupby(by=['area_associata']).sum()
data_with_demo
```

```python
for i in range(len(data_with_demo)):
    for j in range(0, 53):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] / data_with_demo['POP_2013'][i]
    for j in range(53, 106):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] / data_with_demo['POP_2014'][i]
    for j in range(106, 159):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] / data_with_demo['POP_2015'][i]
    for j in range(159, 212):
        data_with_demo.iloc[i,j] = data_with_demo.iloc[i,j] / data_with_demo['POP_2016'][i]

data_with_demo.drop(['POP_2013', 'POP_2014', 'POP_2015', 'POP_2016'], axis=1, inplace=True)
data_with_demo
```

```python
data_with_demo.drop(['SAN COLOMBANO AL LAMBRO'], axis=0, inplace=True)
data_with_demo.shape
```

```python
from scipy.stats import kstest

p_values = []

for i in range(len(data_with_demo)):
    p_values.append(kstest(data_with_demo.iloc[i,0:53], 'norm').pvalue)
    p_values.append(kstest(data_with_demo.iloc[i,53:106], 'norm').pvalue)
    p_values.append(kstest(data_with_demo.iloc[i,106:159], 'norm').pvalue)
    p_values.append(kstest(data_with_demo.iloc[i,159:212], 'norm').pvalue)
    
len(p_values)
```

```python
print(sum(i > 0.05 for i in p_values)  / len(p_values))
```

```python
from scipy.stats import kruskal

p_values_kruskal = []

for i in range(len(data_with_demo)):
    p_values_kruskal.append(kruskal(data_with_demo.iloc[i,0:53], data_with_demo.iloc[i,53:106], data_with_demo.iloc[i,106:159], data_with_demo.iloc[i,159:212]).pvalue)
    
len(p_values_kruskal)
```

```python
print(sum(i > 0.05 for i in p_values_kruskal) / len(p_values_kruskal))   # High percentage!
```

```python
data_checkpoint = data_with_demo.copy()
data_with_demo
```

```python
threshold = 50 * 0.6 * (4/7)

exceeding_2013 = []
exceeding_2014 = []
exceeding_2015 = []
exceeding_2016 = []
intensity_2013 = []
intensity_2014 = []
intensity_2015 = []
intensity_2016 = []

for i in range(len(data_with_demo)):
    temp = sum(data_with_demo.iloc[i,0:53] > threshold)
    exceeding_2013.append(temp)
    temp = sum(data_with_demo.iloc[i,53:106] > threshold)
    exceeding_2014.append(temp)
    temp = sum(data_with_demo.iloc[i,106:159] > threshold)
    exceeding_2015.append(temp)
    temp = sum(data_with_demo.iloc[i,159:212] > threshold)
    exceeding_2016.append(temp)
    
    if ((data_with_demo[data_with_demo > threshold].iloc[i,0:53]).isna().sum()) < 53:
        intensity_2013.append(np.quantile(((data_with_demo[data_with_demo > threshold].iloc[i,0:53] - threshold) / threshold).dropna(), 0.75))
    else:
        intensity_2013.append(0)
    if ((data_with_demo[data_with_demo > threshold].iloc[i,53:106]).isna().sum()) < 53:
        intensity_2014.append(np.quantile(((data_with_demo[data_with_demo > threshold].iloc[i,53:106] - threshold) / threshold).dropna(), 0.75))
    else:
        intensity_2014.append(0)
    if ((data_with_demo[data_with_demo > threshold].iloc[i,106:159]).isna().sum()) < 53:
        intensity_2015.append(np.quantile(((data_with_demo[data_with_demo > threshold].iloc[i,106:159] - threshold) / threshold).dropna(), 0.75))
    else:
        intensity_2015.append(0)
    if ((data_with_demo[data_with_demo > threshold].iloc[i,159:212]).isna().sum()) < 53:
        intensity_2016.append(np.quantile(((data_with_demo[data_with_demo > threshold].iloc[i,159:212] - threshold) / threshold).dropna(), 0.75))
    else:
        intensity_2016.append(0)

df_2013 = pd.DataFrame(data=data_with_demo.index, columns=['area_associata'])
df_2013['Year'] = 2013
df_2013['frequency_PM10'] = exceeding_2013
df_2013['exposure_PM10'] = intensity_2013
df_2013['intensity_PM10'] = df_2013['frequency_PM10'] * df_2013['exposure_PM10']

df_2014 = pd.DataFrame(data=data_with_demo.index, columns=['area_associata'])
df_2014['Year'] = 2014
df_2014['frequency_PM10'] = exceeding_2014
df_2014['exposure_PM10'] = intensity_2014
df_2014['intensity_PM10'] = df_2014['frequency_PM10'] * df_2014['exposure_PM10']

df_2015 = pd.DataFrame(data=data_with_demo.index, columns=['area_associata'])
df_2015['Year'] = 2015
df_2015['frequency_PM10'] = exceeding_2015
df_2015['exposure_PM10'] = intensity_2015
df_2015['intensity_PM10'] = df_2015['frequency_PM10'] * df_2015['exposure_PM10']

df_2016 = pd.DataFrame(data=data_with_demo.index, columns=['area_associata'])
df_2016['Year'] = 2016
df_2016['frequency_PM10'] = exceeding_2016
df_2016['exposure_PM10'] = intensity_2016
df_2016['intensity_PM10'] = df_2016['frequency_PM10'] * df_2016['exposure_PM10']

final_dataset = pd.concat([df_2013, df_2014, df_2015, df_2016], ignore_index=True)
```

```python
final_dataset = final_dataset.round(3)
final_dataset
```

```python
final_dataset.to_csv('Final_datasets_pollution/Final_dataset_PM10.csv', index=False)
```
