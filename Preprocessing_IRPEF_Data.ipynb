```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

def preprocessing_eco(filepath):
    
    data = pd.read_excel(filepath)
    # Remove the years 2020 and 2021 ad split the dataset
    Contr_per_classe = data.iloc[2:10,1:-2].transpose().reset_index()
    Contr_per_classe.columns=['Year', 'Minore o uguale a 0 euro', '0 - 10.000 euro', '10.000 - 15.000 euro', '15.000 - 26.000 euro', '26.000 - 55.000 euro', '55.000 - 75.000 euro','75.000 - 120.000 euro', '120.000 euro e più']
    Redd_per_classe = data.iloc[11:,1:-2].transpose().reset_index()
    Redd_per_classe.columns=['Year', 'Minore o uguale a 0 euro', '0 - 10.000 euro', '10.000 - 15.000 euro', '15.000 - 26.000 euro', '26.000 - 55.000 euro', '55.000 - 75.000 euro', '75.000 - 120.000 euro', '120.000 euro e più']
    
    # Add 2011 data using linear regression
    Contr_2011 = []
    Redd_2011 = []
    x = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]).reshape(-1, 1)
    for i in range(8):
        y = Contr_per_classe.astype('int64').iloc[:,i+1]
        model = LinearRegression().fit(x, y)
        y_pred = model.predict(np.array([2011]).reshape(-1,1))[0]
        Contr_2011.append(round(y_pred))
    for i in range(8):
        y = Redd_per_classe.astype('int64').iloc[:,i+1]
        model = LinearRegression().fit(x, y)
        y_pred = model.predict(np.array([2011]).reshape(-1,1))[0]
        Redd_2011.append(round(y_pred))
    
    Contr_2011.insert(0, 2011)
    Contr_per_classe.loc[-1] = Contr_2011  # adding a row
    Contr_per_classe.index = Contr_per_classe.index + 1  # shifting index
    Contr_per_classe = Contr_per_classe.sort_index()
    Redd_2011.insert(0, 2011)
    Redd_per_classe.loc[-1] = Redd_2011  # adding a row
    Redd_per_classe.index = Redd_per_classe.index + 1  # shifting index
    Redd_per_classe = Redd_per_classe.sort_index()
    
    # Contribuents per income classs
    data_income_classes = np.zeros((9,4))
    data_income_classes[:,0] = Contr_per_classe.iloc[:, 1:3].sum(axis=1)
    data_income_classes[:,1] = Contr_per_classe.iloc[:, 3:5].sum(axis=1)
    data_income_classes[:,2] = Contr_per_classe.iloc[:, 5:7].sum(axis=1)
    data_income_classes[:,3] = Contr_per_classe.iloc[:, 7:9].sum(axis=1)
    data_income_classes = data_income_classes.astype('int64')
    income_classes_df = pd.DataFrame(data=data_income_classes, columns=['I1', 'I2', 'I3', 'I4'])
    income_classes_df['Year'] = Redd_per_classe['Year']
    income_classes_df['Comune'] = filepath.split(sep='.')[0]
    
    # Income statistics
    average_income = np.zeros((9,8))
    Redd_per_classe_f = Redd_per_classe.astype('int64')
    Contr_per_classe_f = Contr_per_classe.astype('int64')

    for i in range(average_income.shape[0]):
        for j in range(average_income.shape[1]):
            if Contr_per_classe_f.iloc[i,j+1] == 0:
                average_income[i,j] = 0
            else:
                average_income[i,j] = Redd_per_classe_f.iloc[i,j+1]/Contr_per_classe_f.iloc[i,j+1]
    statistics_data = np.zeros((9,3))
    statistics_data[:,0] = np.median(average_income, axis=1)
    statistics_data[:,1] = np.quantile(average_income, 0.25, axis=1)
    statistics_data[:,2] = np.quantile(average_income, 0.75, axis=1)
    
    income_classes_df['IMED'] = statistics_data[:,0]
    income_classes_df['IFQ'] = statistics_data[:,1]
    income_classes_df['ITQ'] = statistics_data[:,2]
    
    # Percentages
    pix = np.zeros((9,4))
    pix[:,0] = income_classes_df.iloc[:,0] / income_classes_df.iloc[:,:4].sum(axis=1)
    pix[:,1] = income_classes_df.iloc[:,1] / income_classes_df.iloc[:,:4].sum(axis=1)
    pix[:,2] = income_classes_df.iloc[:,2] / income_classes_df.iloc[:,:4].sum(axis=1)
    pix[:,3] = income_classes_df.iloc[:,3] / income_classes_df.iloc[:,:4].sum(axis=1)
    income_classes_df['PI1'] = pix[:,0]
    income_classes_df['PI2'] = pix[:,1]
    income_classes_df['PI3'] = pix[:,2]
    income_classes_df['PI4'] = pix[:,3]
    
    pit = np.zeros((9,4))
    pit[:,0] = (Redd_per_classe_f.iloc[:,1] + Redd_per_classe_f.iloc[:,2]) / Redd_per_classe_f.iloc[:,1:].sum(axis=1)
    pit[:,1] = (Redd_per_classe_f.iloc[:,3] + Redd_per_classe_f.iloc[:,4]) / Redd_per_classe_f.iloc[:,1:].sum(axis=1)
    pit[:,2] = (Redd_per_classe_f.iloc[:,5] + Redd_per_classe_f.iloc[:,6]) / Redd_per_classe_f.iloc[:,1:].sum(axis=1)
    pit[:,3] = (Redd_per_classe_f.iloc[:,7] + Redd_per_classe_f.iloc[:,8]) / Redd_per_classe_f.iloc[:,1:].sum(axis=1)
    income_classes_df['TI1'] = pit[:,0]
    income_classes_df['TI2'] = pit[:,1]
    income_classes_df['TI3'] = pit[:,2]
    income_classes_df['TI4'] = pit[:,3]
    
    mix = np.zeros((9,4))
    mix[:,0] = (Redd_per_classe_f.iloc[:,1] + Redd_per_classe_f.iloc[:,2]) / income_classes_df.iloc[:,0]
    mix[:,1] = (Redd_per_classe_f.iloc[:,3] + Redd_per_classe_f.iloc[:,4]) / income_classes_df.iloc[:,1]
    mix[:,2] = (Redd_per_classe_f.iloc[:,5] + Redd_per_classe_f.iloc[:,6]) / income_classes_df.iloc[:,2]
    mix[:,3] = (Redd_per_classe_f.iloc[:,7] + Redd_per_classe_f.iloc[:,8]) / income_classes_df.iloc[:,3]
    income_classes_df['MI1'] = mix[:,0]
    income_classes_df['MI2'] = mix[:,1]
    income_classes_df['MI3'] = mix[:,2]
    income_classes_df['MI4'] = mix[:,3]

    income_classes_df[['IMED', 'IFQ', 'ITQ', 'MI1', 'MI2', 'MI3', 'MI4']] = income_classes_df[['IMED', 
    'IFQ', 'ITQ', 'MI1', 'MI2', 'MI3', 'MI4']].apply(lambda x: round(x, 2))
    
    return income_classes_df
```
